```python
from transformers import AutoTokenizer
import transformers, tokenizers

model_name = "microsoft/mdeberta-v3-base"

tok = AutoTokenizer.from_pretrained(model_name)

print("=== VERSION INFO ===")
print("transformers:", transformers.__version__)
print("tokenizers:", tokenizers.__version__)

print("\n=== TOKENIZER TYPE ===")
print(type(tok))
print("is_fast:", tok.is_fast)

print("\n=== SPECIAL TOKENS ===")
print("CLS token:", tok.cls_token_id)
print("SEP token:", tok.sep_token_id)
print("BOS token:", tok.bos_token_id)
print("EOS token:", tok.eos_token_id)

print("\n=== BEHAVIOR TEST ===")
text = "add note ml assignment"
print("add_special_tokens=True :", tok(text, add_special_tokens=True)["input_ids"])
print("add_special_tokens=False:", tok(text, add_special_tokens=False)["input_ids"])

print("\n=== VOCAB INFO ===")
print("vocab_size:", tok.vocab_size)

print("\n=== BACKEND ===")
print(tok.backend_tokenizer)
```
